```python
import requests
from bs4 import BeautifulSoup

import pandas as pd

import re
import time
import math
```

```python
# df 만들기

for pages in range(1,tot_page+1):
    date= "2024.04.09"  ### 페이지별 url 요청
    url = f"https://dart.fss.or.kr/dsac001/mainY.do?selectDate={date}&sort=&series=&mdayCnt=0&currentPage={pages}"
    
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "html.parser")


    ## 전체 페이지와 한 페이지당 건수
    tot_cnt_s = re.findall(r"\d+건",soup.find_all("div", {"class":"pageInfo"})[0].text)[0]
    tot_cnt = re.findall(r"\d+",tot_cnt_s)[0]
    page_lst = soup.find("tbody").find_all("tr")
    
    tot_page = math.ceil(int(tot_cnt)/len(page_lst))
    tot_page


    ## 한 줄 당 데이터
    data_lst = []
    
    for i in range(len(page_lst)):
        temp = soup.find("tbody").find_all("tr")[i]
        sub_time =temp.find_all("td")[0].text.strip()
        comp = temp.find_all("td")[1].find("a").text.strip()
        report = temp.find_all("td")[2].text.strip()
        sub_comp = temp.find_all("td")[3].text
        sub_date = temp.find_all("td")[4].text
    
        data_lst.append([sub_time, comp, report, sub_comp, sub_date])

    #데이터 프레임 생성
    if pages == 1:
        df = pd.DataFrame(data= data_lst,
                         columns =["sub_time", "comp", "report","sub_comp", "sub_date"])

    else :
        df1 = pd.DataFrame(data= data_lst,
                         columns =["sub_time", "comp", "report","sub_comp", "sub_date"])
        df = pd.concat([df, df1])

df
    
```
