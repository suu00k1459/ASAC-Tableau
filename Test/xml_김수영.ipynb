### [데이터 불러오기]

```python
import pandas as pd 
import urllib.request 
from bs4 import BeautifulSoup
```

```python
# step1) url 생성

url_p1 = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.xml"
key = "68195af132b8cc46b7dc02955585c26f"

url_p2 = "40"

url = url_p1 + "?key=" + key+ "&itemPerPage="+url_p2
url
```

```python
# step2) url - request

res = urllib.request.urlopen(url)
soup = BeautifulSoup(res, "html.parser")
```

### [데이터 프레임 만들기]

```python
# 리스트에 정리
movie_list = []

for idx, data in enumerate(soup.find_all("movie")):
    code = data.find("moviecd").text
    name = data.find("movienm").text
    ename = data.find("movienmen").text
    date = data.find("opendt").text
    if data.find_all("peoplenm") !=[]:
        dir_name = data.find("peoplenm").text
    else:
        dir_name = "X" 
        
    movie_list.append([code, name, ename, date, dir_name])


# DF 만들기
df = pd.DataFrame(movie_list, columns = ["code", "Title", "E-Title", "openDay", "Dir_name"])
df
```

```python
# 2. openday의 컬럼을 보고 year, month, day 컬럼 생성

df["year"] = df.loc[:,"openDay"].apply(lambda x : x[:4])
df["month"] = df.loc[:,"openDay"].apply(lambda x : x[4:6])
df["day"] = df.loc[:,"openDay"].apply(lambda x : x[-2:])
df
```

```python
df["genre"]=""
df["ActNum"] = 0
df["ActName"] = ""
df["ActCast"] = ""
```

```python
# 3. 각 영화별 상세 영화 정보 요청해서 xml
#  1번 장르, 배우 수, 1번 배우 이름, 1번 배역이름을 담은 df에다가 추가하기
#  xml 양식으로 처리

url_p1_2 = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.xml"

#### test 
# url_p1_2 + "?key=" + key + "&movieCd="+"20234045"
# res2 = urllib.request.urlopen(url2)
# soup2 = BeautifulSoup(res2, "html.parser")
# temp = soup2.find("movieinforesult").find("movieinfo")
# print(temp)



for i in df.index:
    url2 = url_p1_2 + "?key=" + key + "&movieCd="+i
    res2 = urllib.request.urlopen(url2)
    
    soup2 = BeautifulSoup(res2, "html.parser")
    
    temp = soup2.find("movieinforesult").find("movieinfo")

    df.loc[i,"genrenm"] = temp.find("genrenm").text
    df.loc[i,"ActNum"] = len(temp.find("actors"))

    if temp.find_all("actor") !=[]:
        df.loc[i,"ActName"] = temp.find("actor").find_all("peoplenm")[0].text 
        df.loc[i,"ActCast"] = temp.find("actor").find("cast").text
    else:
        df.loc[i,"ActName"] =""
        df.loc[i,"ActCast"] = ""

       
```
